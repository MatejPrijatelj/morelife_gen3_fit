```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.interpolate import UnivariateSpline

# Load your data (replace with your actual data)
# Example: x = np.array([...]), y = np.array([...])
x = np.linspace(0, 10, 20)
y = np.log1p(x) + np.random.normal(0, 0.1, size=len(x))  # Example smooth dataset

# Define potential non-oscillatory functions
def exp_func(x, a, b, c):
    return a * np.exp(b * x) + c

def poly_func(x, a, b, c):
    return a*x**2 + b*x + c  # Low-degree polynomial

# Fit exponential function
popt_exp, _ = curve_fit(exp_func, x, y, maxfev=5000)

# Fit polynomial (quadratic)
popt_poly = np.polyfit(x, y, 2)

# Fit a smoothing spline
spline = UnivariateSpline(x, y, s=1)

# Generate smooth x values
x_fit = np.linspace(min(x), max(x), 100)
y_exp_fit = exp_func(x_fit, *popt_exp)
y_poly_fit = np.polyval(popt_poly, x_fit)
y_spline_fit = spline(x_fit)

# Plot results
plt.scatter(x, y, label="Data", color='black')
plt.plot(x_fit, y_exp_fit, label="Exponential Fit", linestyle='dashed')
plt.plot(x_fit, y_poly_fit, label="Polynomial Fit", linestyle='dotted')
plt.plot(x_fit, y_spline_fit, label="Spline Fit", linestyle='dashdot')
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Non-Oscillatory Fitting")
plt.show()

```

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import solve

# Generate noisy data (Example)
np.random.seed(0)
x = np.linspace(0, 10, 30)
y = np.sin(x) + 0.5 * np.random.randn(len(x))  # Noisy sine wave

# Construct the regularized least squares fit
degree = 5  # Degree of polynomial (keep low to avoid oscillations)
lambda_penalty = 100  # Smoothness penalty (higher -> smoother)

# Design matrix for polynomial fitting (Vandermonde matrix)
A = np.vander(x, degree + 1)

# Compute the second derivative penalty matrix (finite difference approximation)
D2 = np.diag(np.ones(degree - 1), -1) - 2 * np.diag(np.ones(degree)) + np.diag(np.ones(degree - 1), 1)
D2 = D2[1:-1, :]  # Trim it to match the size of the polynomial degree

# Solve for coefficients with the regularization (penalty)
ATA = A.T @ A
ATy = A.T @ y
D2TD2 = D2.T @ D2  # Smoothness penalty term
coefs = solve(ATA + lambda_penalty * D2TD2, ATy)

# Evaluate the fitted polynomial at a dense grid of x values
x_fit = np.linspace(0, 10, 100)
y_fit = np.polyval(coefs, x_fit)

# Plot the original data and the regularized fit
plt.scatter(x, y, label="Noisy Data", color="black")
plt.plot(x_fit, y_fit, label="Smooth Regularized Fit", color="red", linewidth=2)
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Oscillation-Penalized Polynomial Fit (Regularized)")
plt.show()
```
